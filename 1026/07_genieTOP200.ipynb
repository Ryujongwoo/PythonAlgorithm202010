```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
import time
header = {'User-agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko'}
```

```python
# 타겟 사이트 주소 만들기
url = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20201026&hh=19&rtm=Y&pg=' # 변경되지 않는 주소
for i in range(1, 5):
    # print(url + str(i))
    print('{}{}'.format(url, i))
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20201026&hh=19&rtm=Y&pg=1 =>   1 ~  50
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20201026&hh=19&rtm=Y&pg=2 =>  51 ~ 100
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20201026&hh=19&rtm=Y&pg=3 => 101 ~ 150
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20201026&hh=19&rtm=Y&pg=4 => 151 ~ 200
```

```python
# 노래 제목 크롤링
for i in range(1, 2):
    targetSite = '{}{}'.format(url, i)
    # print(targetSite)
    request = requests.get(targetSite, headers = header)
    time.sleep(1)
    # print(request)
    html = request.text
    # print(html)
    soup = BeautifulSoup(html, 'html.parser')
    # print(soup)
    
    titles = soup.findAll('a', {'class': 'title ellipsis'})
    # print(titles)
    # print(len(titles))
    for title in titles:
        # print(title.text.strip())
        # print(title.text.split('\n')[-1])
        print(title.text.strip().split('\n')[0])
```

```python
# 아티스트 크롤링
for i in range(1, 2):
    targetSite = '{}{}'.format(url, i)
    request = requests.get(targetSite, headers = header)
    time.sleep(1)
    html = request.text
    soup = BeautifulSoup(html, 'html.parser')
    
    artists = soup.findAll('a', {'class': 'artist ellipsis'})
    for artist in artists[5:]:
        print(artist.text.strip().split('\n')[0])
```

```python
genies = []
for i in range(1, 5):
    targetSite = '{}{}'.format(url, i)
    request = requests.get(targetSite, headers = header)
    time.sleep(1)
    html = request.text
    soup = BeautifulSoup(html, 'html.parser')
    titles = soup.findAll('a', {'class': 'title ellipsis'})
    artists = soup.findAll('a', {'class': 'artist ellipsis'})
    for i in range(len(titles)):
        genies.append([titles[i].text.strip().split('\n')[0], artists[i + 5].text.strip().split('\n')[0]])
```

```python
print(dt.now())
for i in range(len(genies)):
    print('{0:3d}위. {1} - {2}'.format(i + 1, genies[i][0], genies[i][1]))
```

```python
file = open('genieTOP200.txt', 'w')
file.write('{} 현재 Genie 뮤직 실시간 TOP 200\n'.format(dt.now()))

for i in range(200):
    data = '{0:3d}위 {1} - {2}'.format(i + 1, genies[i][0], genies[i][1])
    file.write(data + '\n')
    
file.close()
print('genieTOP200.txt 파일로 쓰기 완료')
```
