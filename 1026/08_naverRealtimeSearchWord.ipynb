```python
#     
# 2020 3       .
#        ajax    .
# requests  get()     json  .
targetSite = 'https://apis.naver.com/mobile_main/srchrank/srchrank?frm=main&ag=40s&gr=4&ma=-2&si=2&en=-2&sp=2'
header = {'User-agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko'}
Image('./naver2.png', width = '1200')
```

```python
request = requests.get(targetSite, headers = header)
word = request.text
print(type(word))
print(word)
```

```python
# json       json  import .
import json
```

```python
# json  loads()   json    .
word_json = json.loads(word)
print(type(word_json))
print(word_json)
```

```python
#    .
# ranks = word_json.get('data')
ranks = word_json['data']
print(type(ranks))
print(ranks)
```

```python
#           .
for rank in ranks:
    print('{0:2d} : {1}'.format(rank['rank'], rank['keyword']))
```
